scale the data to increase the acuricy in this case we should device our whole data set by 255
<h3>Why divide by 255?</h3>

<p>In the MNIST dataset, pixel values range from <code>0</code> to <code>255</code> (uint8 format).</p>

<p>Dividing by 255 scales all values to the range <code>[0, 1]</code>.</p>

<p>This makes the data easier for neural networks to learn from — it helps with:</p>

<ul>
  <li>Faster convergence during training</li>
  <li>More stable gradients</li>
  <li>Better performance</li>
</ul>


```python
X_train = X_train/255
X_test = X_test/255
```

now it is time to change the 3 dimensions arry to 2 d_array by calling reshape function 

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
#this is after changes to 2D array and it is 2D array why we change because every element(pixcel) is one neuron
X_train_flattened.shape
```

```python
#it is before reshapeing and it is 3D rgigh now
X_train.shape
```

```python
#at first we train our model witout hidden layer and then we train another model with hidden layer and compare their accuricy
#Sequential means that we have one stack of layers, Dense means to interconnection between neurons 
#output is 10 because we 0 - 9 ten neurons and input is 784 because we 28*28 
model = keras.Sequential([
    keras.layers.Dense(10,input_shape = (784,), activation='sigmoid')
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
history = model.fit(X_train_flattened, y_train, epochs = 7)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
# Plot training accuracy
plt.plot(history.history['accuracy'], label='Accuracy')
# Plot training loss
plt.plot(history.history['loss'], label='Loss')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Training Accuracy and Loss per Epoch')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_test_predicted = model.predict(X_test_flattened)
np.argmax(X_test_predicted[0])
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted_label = [np.argmax(i) for i in X_test_predicted]
y_predicted_label[:5]
```

now by using confusion matrix we can check where our model got errors or got confusion 

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_label)
cm
```

```python
i   mport seaborn as sn

plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted') 
plt.ylabel('Truth')
plt.title('Confusion matrix')
```

now let's add a hidden layer to improve the performance 

```python
model_h = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_h.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_h.fit(X_train_flattened, y_train, epochs = 5)
```

```python
#model with hidden layer 0.97
model_h.evaluate(X_test_flattened, y_test)
```

```python
#accuricy of our model without of hidden layer  0.92
model.evaluate(X_test_flattened, y_test)
```

```python
model_f = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_f.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_f.fit(X_train, y_train, epochs = 5)
```

```python
model_f.evaluate(X_test, y_test)
```
